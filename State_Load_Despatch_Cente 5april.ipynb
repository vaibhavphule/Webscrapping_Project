## 1.  Required data from 01-01-2021 to till yesterday your working day
## 2.  Save all the data in individual files for each days you can store in any format

```python
import datetime
base = datetime.datetime.today()
t=base-datetime.datetime(2021, 1,1)
date_list = [base - datetime.timedelta(days=x) for x in range(t.days+1)]
dates=[d.strftime("%d-%m-%Y") for d in date_list] 
dates.reverse()
dates
```

```python
today=base-datetime.datetime(2021, 1,1)
today.days
```

```python
for d in dates:
    d1=d.replace('-','/')
    url=f"https://www.delhisldc.org/Loaddata.aspx?mode={d1}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'lxml')
    #print(soup)
    data=soup.find(id="ContentPlaceHolder3_DGGridAv")
    rows=data.find_all('tr')
    heads=[p.get_text() for p in rows[0].find_all('td')]
    table=[]
    for d in rows[1:]:
        table.append([p.get_text() for p in d.find_all('td')])
    dict={}
    for k in heads:
        dict[k]=[]
    for t in table:
        dict['TIMESLOT'].append(t[0])
        dict['DELHI'].append(t[1])
        dict['BRPL'].append(t[2])
        dict['BYPL'].append(t[3])
        dict['NDPL'].append(t[4])
        dict['NDMC'].append(t[5])
        dict['MES'].append(t[6])
    df=pd.DataFrame(dict)
    r=d1.replace('/','_')
    print(r)
    df.to_csv(f'data/{r}.csv',index=False)
```
