```python
import pandas as pd
d={'titles':[],'authors':[],'prizes':[]}
for i in range(1,35,1):
    url=f"https://www.bookdepository.com/bestsellers?page={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html5lib')
    all_book=soup.find_all(class_="book-item")
    for f in all_book:
        d['titles'].append(f.find(class_="title").get_text().strip())
        d['authors'].append(f.find(class_="author").get_text().strip())
        d['prizes'].append(f.find(class_="price-wrap").get_text().strip().split('\n')[0].lstrip('₹'))
df=pd.DataFrame(d)
df.head(30)
    
```
