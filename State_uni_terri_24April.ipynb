```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/States_and_union_territories_of_India"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
#soup
```

```python
data=soup.find('table',{"class":"wikitable sortable plainrowheaders"})
data
```

```python
heads=data.find_all('th')
heading=[h.get_text().strip() for h in heads][:11]
heading
```

```python

States=[h.get_text().strip() for h in heads][11:]
States
```

```python
table_body=data.find_all('tr')[1:]
table_body
#len(table_body)
```

```python
first=[d.get_text().strip()  for d in table_body[0].find_all('td')]
first
```

```python
comp=[]
for t in table_body:
    l=[d.get_text().strip()  for d in t.find_all('td')]
    comp.append(l)
len(comp)
```

```python
comp
```

```python
d={}
for h in heading:
    d[h]=[]
d
```

```python
for k in comp:
    if len(k)==9:
        d['ISO 3166-2:IN'].append(k[0])
        d['Vehiclecode'].append(k[1])
        d['Zone'].append(k[2])
        d['Capital'].append(k[3])
        d['Largest city'].append(k[3])
        d['Statehood'].append(k[4])
        d['Population(census 2011)[citation needed]'].append(k[5])
        d['Area(km2)'].append(k[6])
        d['Officiallanguages[39]'].append(k[7])
        d['Additional officiallanguages[39]'].append(k[8])
    else:
        d['ISO 3166-2:IN'].append(k[0])
        d['Vehiclecode'].append(k[1])
        d['Zone'].append(k[2])
        d['Capital'].append(k[3])
        d['Largest city'].append(k[4])
        d['Statehood'].append(k[5])
        d['Population(census 2011)[citation needed]'].append(k[6])
        d['Area(km2)'].append(k[7])
        d['Officiallanguages[39]'].append(k[8])
        d['Additional officiallanguages[39]'].append(k[9])
d
        
```
