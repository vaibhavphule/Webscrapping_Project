```python

d={'Questions':[],'Votes':[],'views':[]}
for i in range(1,6,1):
    url=f"https://stackoverflow.com/questions?tab=newest&page={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html5lib')
    all_ques=soup.find(class_="flush-left")
    ques_list=all_ques.find_all(class_="question-summary") 
    for f in ques_list:
        d['Questions'].append(f.find(class_="excerpt").get_text().strip().replace('\n',''))
        d['Votes'].append(f.find(class_="vote-count-post").get_text())
        d['views'].append(f.find(class_="views").get_text().strip().replace(' views',''))
df=pd.DataFrame(d)
df.head(50)
```

```python
df.to_csv("stackoverflow.csv",index=False)
```
