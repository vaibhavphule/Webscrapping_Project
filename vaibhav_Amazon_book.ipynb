```python
#https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=25
import pandas as pd
d={'titles':[],'author':[],'prizes':[],'Stars':[]}
for i in range(1,3,1):
    url=f"https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html5lib')
    all_book=soup.find_all(class_="zg-item-immersion") 
    for f in all_book:
        if f.find(class_="a-size-small a-link-child")!=None:
            d['titles'].append(f.find(class_="a-link-normal").get_text().strip())
            d['author'].append(f.find(class_="a-size-small a-link-child").get_text().strip())
            d['prizes'].append(f.find(class_="p13n-sc-price").get_text().strip().split('\n')[0].strip('₹\xa0'))
            d['Stars'].append(f.find(class_="a-icon-alt").get_text().strip())
        else:
            d['titles'].append(f.find(class_="a-link-normal").get_text().strip())
            d['author'].append(['author not avaible'])
            d['prizes'].append(f.find(class_="p13n-sc-price").get_text().strip().split('\n')[0].strip('₹\xa0'))
            d['Stars'].append(f.find(class_="a-icon-alt").get_text().strip())
df=pd.DataFrame(d)
df.head(50)
```
